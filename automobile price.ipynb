```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
data = pd.read_csv('data/Automobile price data.csv')
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
cols = ['price', 'bore', 'stroke', 'horsepower', 'peak-rpm']
for column in cols:
    data.loc[data[column] == '?', column] = np.nan
data.dropna(axis = 0, inplace = True)
```

```python
for column in cols:
    data[column] = pd.to_numeric(data[column])
```

```python
plt.plot(data['city-mpg'], data['price'], 'ro')
plt.xlabel('city-mpg')
plt.ylabel('price')
plt.title('city-mpg vs price')
```

expensive cars consume more fuels, while cheap cars are economical

```python
fig = plt.figure(figsize = (6, 6))
ax = fig.gca()
data.plot(kind = 'scatter', x = 'curb-weight', y='price',  ax = ax, color = 'g')
ax.set_xlabel('Curb weight')
ax.set_ylabel('price')
ax.set_title('curb weight vs price')
```

Heavy cars are expensive, while lighter cars are cheap

```python
cars = data['make'].value_counts()
```

```python
fig = plt.figure(figsize= (8, 8))
ax = fig.gca()
cars.plot(kind = 'bar', ax = ax)
ax.set_title('car brands')
ax.set_xlabel('brand')
ax.set_ylabel('number')
```

Toyota is the preferred car 

```python
fig = plt.figure(figsize= (8, 8))
ax = fig.gca()
cars.plot(kind = 'hist', ax = ax)
ax.set_title('car brands')
ax.set_xlabel('brand')
ax.set_ylabel('number')
```

```python
fig = plt.figure(figsize = (6, 8))
ax = fig.gca()
data[['engine-size','fuel-type']].boxplot(by = 'fuel-type', ax = ax)
ax.set_title('engine-size and fuel type')
ax.set_xlabel(' fuel type')
ax.set_ylabel('engine size')
```

the median size for gas car is lower than diesel car

```python
fig = plt.figure(figsize = (6, 6))
ax = fig.gca()
sns.set_style('dark')
sns.kdeplot(data['city-mpg'], ax = ax)
ax.set_title('KDE plot of auto engine size') 
ax.set_xlabel('Engine Size') 
ax.set_ylabel('Density')
```

few outliers and the engine size is skewed towards smaller cars

```python
fig = plt.figure(figsize = (6, 6))
ax = fig.gca()
sns.set_style('whitegrid')
sns.kdeplot(data[['city-mpg', 'engine-size', 'price']], ax = ax, cmap="Blues_d")
ax.set_title('KDE plot of auto engine size') 
ax.set_xlabel('city mpg') 
ax.set_ylabel('price')
```

```python
fig = plt.figure(figsize = (6, 6))
ax = fig.gca()
sns.set_style('whitegrid')
sns.kdeplot(data[[ 'engine-size', 'price']], ax = ax, cmap="Blues_d")
ax.set_title('KDE plot of auto engine size') 
ax.set_xlabel('city mpg') 
ax.set_ylabel('price')
```

```python
fig = plt.figure(figsize=(8,8)) 
ax = fig.gca() 
sns.set_style("whitegrid")
sns.violinplot(x = 'fuel-type', y = 'engine-size', data = data, ax = ax)
ax.set_title('Violine plots of auto engine size by fuel type') 
ax.set_xlabel('Fuel type')
ax.set_ylabel('Engine size')
```
